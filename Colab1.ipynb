```python
#quiz 1 ให้เลือกเฉพาะคอลัม 'id', 'name', จำนวนรีวิว, ประเภทร้าน, จำนวนเชคอิน
c = ['id', 'name', 'number_of_reviews', 'category_name', 'number_of_checkins']
x = w_restaurant.loc[:, c]
x.head(10)
```

```python
#เลือกแถว เงื่อนไขจาก แถวที่เป็น ข้อมูลตัวเลข
#ต้องการเฉพาะแถวที่มี weighted_average_rating > 3.0
condition = w_restaurant['weighted_average_rating'] > 3.0
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(5)
```

```python
#quiz2 เลือกเฉพาะ จำนวนรีวิว เกิน 10 (number_of_reviews >10)
condition = w_restaurant['number_of_reviews'] > 10
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(5)
```

```python
#เงื่อนไข คอลัมเป็นข้อมุลประเภมข้อความ
#1. ขึ้นต้นด้วยคำว่า ...
#2. ลงท้ายด้วยคำว่า ...
#3. มีคำว่า ... อยู่ในชื่อ

#ตัวอย่าง name ขึ้นต้นด้วยคำว่า 'ก๋วนเตี๋ยว'
condition = w_restaurant['name'].str.startswith('ก๋วยเตี๋ยว')
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#quiz 3: หาร้านที่อยู่ในจังหวัดที่ขึ้นต้นด้วยคำว่า นคร (Nalhon)
condition = w_restaurant['city_name'].str.startswith('์Nakhon')
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#ตัวอย่าง name ลงท้ายด้วยคำว่า 'ดงมูลเหล็ก'
condition = w_restaurant['name'].str.endswith('์ดงมูลเหล็ก')
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#ชื่อร้านมีคำว่า ขาหมู อยู่ในร้าน
condition = w_restaurant['name'].str.contains('์ขาหมู')
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#หาว่ามีแถวใดที่เป็นค่าว่าง
#ตัวอย่าง หาร้านที่ zip code เป็นค่าว่าง
condition = w_restaurant['zipcode'].isna()
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#หาว่ามีแถวใดที่เป็นค่าไม่ว่าง
#ตัวอย่าง หาร้าน zip code เป็นค่าไม่ว่าง
condition = w_restaurant['zipcode'].notna()
w_restaurant_selection = w_restaurant.loc[condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#การเชื่อมเงื่อนไข and or not
#หาร้านที่มีเงื่อนไข 2 อย่าง
#1. zipcode ไม่ว่าง
#2. และ มีเลข 60 อยู่ใน zipcode
# and &
# or |
# not
condition1 = w_restaurant['zipcode'].notna()
condition2 = w_restaurant['zipcode'].str.contains('60')
condition = condition1 & condition2
w_restaurant_selection = w_restaurant.loc[condition,:]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
condition1 = w_restaurant['weighted_average_rating'] >= 2
condition2 = w_restaurant['weighted_average_rating'] <= 4
condition = condition1 & condition2

select_column = ['id', 'name', 'weighted_average_rating']
w_restaurant_selection = w_restaurant.loc[condition, select_column]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#กานเชื่อมเงื่อนไข and or not
#ตัวอย่าว or
#
condition1 = w_restaurant['name'].str.contains('ขาหมู')
condition2 = w_restaurant['name'].str.contains('ก๋วยเตี๋ยว')
condition = condition1 | condition2
w_restaurant_selection = w_restaurant.loc[condition,:]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#quiz 5
#หาร้านที่อยู่ในกรุงเทพที่เรตติ้ง > 4 หรือ อยู่ในเชียงใหม่ที่เรตติ้ง > 3
#Bangkok Metropolitan Region
#Chiang Mai
print(w_restaurant.columns)
condition_bangkok = (w_restaurant['city_name'] == 'Bangkok') & (w_restaurant['weighted_average_rating'] > 4)
condition_chiangmai = (w_restaurant['city_name'] == 'Chiang Mai') & (w_restaurant['weighted_average_rating'] > 3)
final_condition = condition_bangkok | condition_chiangmai
w_restaurant_selection = w_restaurant.loc[final_condition, :]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```
